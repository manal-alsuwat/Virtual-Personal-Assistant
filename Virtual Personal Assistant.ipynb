```python
import pyttsx3 
import speech_recognition as sr 
import datetime 
import wikipedia 
import webbrowser 
import os 
import ctypes 
 
```

Now, Weâ€™ll be initializing a variable and getting the necessary voice parameter we need. For female voice, you can set the value as 1 in the second line and 0 for male voice. 

```python
engine = pyttsx3.init('sapi5') 
voices = engine.getProperty('voices') 
engine.setProperty('voice', voices[1].id) 
def speak(audio): 
	engine.say(audio) 
	engine.runAndWait()
```
